## THE BUCK (STEP-DOWN) CONVERTER

Controlling the dc component of a pulsed output voltage of the type in Fig. 06002c may be sufficient for some applications, such as controlling the speed of a dc motor, but often the objective is to produce an output that is purely dc. One way of obtaining a dc output from the circuit of Fig. 06002a is to insert a low-pass filter after the switch. Figure 06003a shows an $L C$ low-pass filter added to the basic converter. The diode provides a path for the inductor current when the switch is opened and is reverse-biased when the switch is closed. This circuit is called a buck comverter or a step-down comverter because the output voltage is less than the input.

<center> Figure 06002.$\quad$ (a) A basic dc-dc switching converter; (b) Switching
equivalent; (c) Output voltage. </center>
<br><br/>
<br><br/>

$$
V_{o}=\frac{1}{T} \int_{0}^{T} v_{o}(t) d t=\frac{1}{T} \int_{0}^{D T} V_{s} d t=V_{s} D
$$   [Eq. Eq. 06001]


__Voltage and Current Relationships__
If the low-pass filter is ideal, the output voltage is the average of the input voltage to the filter. The input to the filter, $v_{x}$ in Fig. 06003a, is $V_{s}$ when the switch is closed and is zero when the switch is open, provided that the inductor current remains positive, keeping the diode on. If the switch is closed periodically at a duty ratio $D,$ the average voltage at the filter input is $V_{s} D,$ as in [Eq. 06001]

This analysis assumes that the diode remains forward-biased for the entire time when the switch is open, implying that the inductor current remains positive. __An inductor current that remains positive throughout the switching period is known as continuous current. Conversely, discontinuous current is characterized by the inductor current's returning to zero during each period.__

Another way of analyzing the operation of the buck converter of Fig. 06003a is to examine the inductor voltage and current. This analysis method will prove useful for designing the filter and for analyzing circuits that are presented later in this section.

Buck converters and de-dc converters in general, have the following properties when operating in the steady state:

1. The inductor current is periodic.

$$
i_{L}(t+T)=i_{L}(t)
$$   [Eq. 06003]

2. The average inductor voltage is zero

$$
V_{L}=\frac{1}{T} \int_{t}^{t+T} v_{L}(\lambda) d \lambda=0
$$   [Eq. 06004]

3. The average capacitor current is zero

$$
I_{C}=\frac{1}{T} \int_{t}^{t+T} t_{C}(\lambda) d \lambda=0
$$   [Eq. 06005]

4. The power supplied by the source is the same as the power delivered to the load. For nonideal components, the source also supplies the losses.

$$
\begin{array}{ll}
P_{s}=P_{o} & \text { ideal } \\
P_{s}=P_{o}+\text { losses } & \text { nonideal }
\end{array}
$$   [Eq. 06006]

Analysis of the buck converter of Fig. 06003a begins by making these assumptions:

1. The circuit is operating in the steady state.
2. The inductor current is continuous (always positive).
3. The capacitor is very large, and the output voltage is held constant at voltage $V_{o}$. This restriction will be relaxed later to show the effects of finite capacitance.
4. The switching period is $T$; the switch is closed for time $D T$ and open for time $(1-D) T$
5. The components are ideal.

The key to the analysis for determining the output $V_{o}$ is to examine the inductor current and inductor voltage first for the switch closed and then for the switch open. __The net change in inductor current over one period must be zero for steadystate operation (periodic current singal).__ The average inductor voltage is zero.

__Analysis for the Switch Closed__ When the switch is closed in the buck converter circuit of Fig. 06003a, the diode is reverse-biased and Fig. 06003b is an equivalent circuit. The voltage across the inductor is

$$
v_{L}=V_{s}-V_{o}=L \frac{d i_{L}}{d t}
$$

Rearranging,

$$
\frac{d i_{L}}{d t}=\frac{V_{s}-V_{o}}{L} \quad \text { switch closed }
$$

since the derivative of the current is a positive constant, the current increases linearly as shown in Fig. 06004b. The change in current while the switch is closed is computed by modifying the preceding equation.

$$
\begin{aligned}
\frac{d i_{L}}{d t}=\frac{\Delta i_{L}}{\Delta t} &=\frac{\Delta i_{L}}{D T}=\frac{V_{s}-V_{o}}{L} \\
\left(\Delta i_{L}\right)_{\text {closed }} &=\left(\frac{V_{s}-V_{o}}{L}\right) D T
\end{aligned}
$$   [Eq. 06007]

<br><br/>
<p align="center"><img src="images/buck_converter_waveforms.png" style="width:39%"></p>
<center> Figure 06004.$\quad$ Buck converter waveforms: (a) Inductor voltage; (b) Inductor current; (c) Capacitor current. </center>
<br><br/>
<br><br/>

__Analysis for the Switch Open__ When the switch is open, the diode becomes forward-biased to carry the inductor current and the equivalent circuit of Fig. 06003c applies. The voltage across the inductor when the switch is open is

$$
v_{L}=-V_{o}=L \frac{d i_{L}}{d t}
$$

Rearranging,

$$
\frac{d i_{L}}{d t}=\frac{-V_{o}}{L} \quad \text { switch open }
$$

The derivative of current in the inductor is a negative constant, and the current decreases linearly as shown in Fig. 06004b. The change in inductor current when the switch is open is

$$
\begin{array}{c}
\frac{\Delta i_{L}}{\Delta t}=\frac{\Delta i_{L}}{(1-D) T}=-\frac{V_{o}}{L} \\
\left(\Delta l_{L}\right)_{\text {open }}=-\left(\frac{V_{o}}{L}\right)(1-D) T
\end{array}
$$   [Eq. 06008]

Steady-state operation requires that the inductor current at the end of the switching cycle be the same as that at the beginning, meaning that the net change in inductor current over one period is zero. This requires

$$
\left(\Delta i_{L}\right)_{\text {closed }}+\left(\Delta l_{L}\right)_{\text {open }}=0
$$

Using [Eq. 06007] and [Eq. 06008]

$$
\left(\frac{V_{s}-V_{o}}{L}\right)(D T)-\left(\frac{V_{o}}{L}\right)(1-D) T=0
$$

Solving for $V_{o}$

$$
V_{o}=V_{s} D
$$   [Eq. 06009]

which is the same result as [Eq. 06001]. The buck converter produces an output voltage that is less than or equal to the input.

An alternative derivation of the output voltage is based on the inductor voltage, as shown in Fig. 06004a. Since the average inductor voltage is zero for periodic operation,

$$
V_{L}=\left(V_{s}-V_{o}\right) D T+\left(-V_{o}\right)(1-D) T=0
$$

Solving the preceding equation for $V_{o}$ yields the same result as [Eq. 06009], $V_{o}=V_{s} D$ Note that the output voltage depends on only the input and the duty ratio $D$. If the input voltage fluctuates, the output voltage can be regulated by adjusting the duty ratio appropriately. A feedback loop is required to sample the output voltage, compare it to a reference, and set the duty ratio of the switch accordingly. Regulation techniques are discussed in the section [Digitally Controlled Switched-Mode Converters](https://github.com/Xiuhcoatl-013/Energy-Field-Theory/blob/master/notes/digitally_controlled_switched-mode_converters/digitally_controlled_switched-mode_converters.ipynb).

The average inductor current must be the same as the average current in the load resistor, since the average capacitor current must be zero for steady-state operation:

$$
I_{L}=I_{R}=\frac{V_{o}}{R}
$$   [Eq. 06010]

since the change in inductor current is known from [Eqs. 06007] and [Eqs. 06008], the maximum and minimum values of the inductor current are computed as

$$
\begin{aligned}
I_{\text {max }} &=I_{L}+\frac{\Delta i_{L}}{2} \\
&=\frac{V_{o}}{R}+\frac{1}{2}\left[\frac{V_{o}}{L}(1-D) T\right]=V_{o}\left(\frac{1}{R}+\frac{1-D}{2 L f}\right)
\end{aligned}
$$   [Eq. 06011]

$$
\begin{aligned}
I_{\text {min }} &=I_{L}-\frac{\Delta i_{L}}{2} \\
&=\frac{V_{o}}{R}-\frac{1}{2}\left[\frac{V_{o}}{L}(1-D) T\right]=V_{o}\left(\frac{1}{R}-\frac{1-D}{2 L f}\right)
\end{aligned}
$$   [Eq. 06012]

where $f = 1/T$ is the switching frequency.

For the preceding analysis to be valid, continuous current in the inductor must be verified. An easy check for continuous current is to calculate the minimum inductor current from [Eq. 06012]. Since the minimum value of inductor current must be positive for continuous current, a negative minimum calculated from [Eq. 06012] is not allowed due to the diode and indicates discontinuous current. The circuit will operate for discontinuous inductor current, but the preceding analysis is not valid. Discontinuous-current operation is discussed later in this section.

The [Eq. 06012] can be used to determine the combination of $L$ and $f$ that will result in continuous current. since $I_{\min }=0$ is the boundary between continuous and discontinuous current,

$$
\begin{array}{c}
I_{\min }=0=V_{o}\left(\frac{1}{R}-\frac{1-D}{2 L f}\right) \\
(L f)_{\min }=\frac{(1-D) R}{2}
\end{array}
$$   [Eq. 06013]

If the desired switching frequency is established,

$$
L_{\min }=\frac{(1-D) R}{2 f} \quad \text { for continuous current }
$$   [Eq. 06014]

where $L_{\min }$ is the minimum inductance required for continuous current. In practice, a value of inductance greater than $L_{\min }$ is desirable to ensure continuous current.

In the design of a buck converter, the peak-to-peak variation in the inductor current is often used as a design criterion. [Eq. 06007] can be combined with [Eq. 06009] to determine the value of inductance for a specified peak-to-peak inductor current for continuous-current operation:

$$
\Delta i_{L}=\left(\frac{V_{s}-V_{o}}{L}\right) D T=\left(\frac{V_{s}-V_{o}}{L f}\right) D=\frac{V_{s}(1-D)}{L f}
$$   [Eq. 06015]

or

$$
L=\left(\frac{V_{s}-V_{o}}{\Delta i_{L} f}\right) D=\frac{V_{o}(1-D)}{\Delta i_{L} f}
$$   [Eq. 06016]

since the converter components are assumed to be ideal, the power supplied by the source must be the same as the power absorbed by the load resistor.

$$
V_{s} I_{s}=V_{o} I_{o}
$$

$$V_{s} I_{s}=V_{o} I_{o}$$  [Eq. 06017]

or

$$
\frac{V_{o}}{V_{s}}=\frac{I_{s}}{I_{o}}
$$

Note that the preceding relationship is similar to the voltage-current relationship for a transformer in ac applications. Therefore, ___the buck converter circuit is equivalent to a dc transformer.___

__Output Voltage Ripple__
In the preceding analysis, the capacitor was assumed to be very large to keep the output voltage constant. In practice, the output voltage cannot be kept perfectly constant with a finite capacitance. The variation in output voltage, or ripple, is computed from the voltage-current relationship of the capacitor. The current in the capacitor is

$$
i_{C}=i_{L}-i_{R}
$$

shown in [Fig. 06005a] While the capacitor current is positive, the capacitor is charging. From the definition of capacitance,

$$
\begin{aligned}
Q &=C V_{o} \\
\Delta Q &=C \Delta V_{o} \\
\Delta V_{o} &=\frac{\Delta Q}{C}
\end{aligned}
$$

The change in charge $\Delta Q$ is the area of the triangle above the time axis

$$
\Delta Q=\frac{1}{2}\left(\frac{T}{2}\right)\left(\frac{\Delta i_{L}}{2}\right)=\frac{T \Delta i_{L}}{8}
$$

resulting in

$$
\Delta V_{o}=\frac{T \Delta i_{L}}{8 C}
$$

<br><br/>
<p align="center"><img src="images/buck_converter_output_voltage_ripple_waveforms.png" style="width:39%"></p>
<center> Figure 06005.$\quad$ Buck converter waveforms. (a) Capacitor current; (b) Capacitor ripple voltage. </center>
<br><br/>
<br><br/>

Using [Eq. 06008] for $\Delta i_{L}$

$$
\Delta V_{o}=\frac{T V_{o}}{8 C L}(1-D) T=\frac{V_{o}(1-D)}{8 L C f^{2}}
$$   [Eq. 06018]

In this equation, $\Delta V_{o}$ is the peak-to-peak ripple voltage at the output, as shown in Fig. 06005b. It is also useful to express the ripple as a fraction of the output voltage,

$$
\frac{\Delta V_{o}}{V_{o}}=\frac{1-D}{8 L C f^{2}}
$$   [Eq. 06019]

In design, it is useful to rearrange the preceding equation to express required capacitance in terms of specified voltage ripple:

$$
C=\frac{1-D}{8 L\left(\Delta V_{o} / V_{o}\right) f^{2}}
$$   [Eq. 06020]

If the ripple is not large, the assumption of a constant output voltage is reasonable and the preceding analysis is essentially valid.

__Capacitor Resistance-The Effect on Ripple Voltage__
The output voltage ripple in [Eq. 06018] and [Eq. 06019] is based on an ideal capacitor. A real capacitor can be modeled as a capacitance with an equivalent series resistance (ESR) and an equivalent series inductance (ESL). The ESR may have a significant effect on the output voltage ripple, often producing a ripple voltage greater than that of the ideal capacitance. The inductance in the capacitor is usually not a significant factor at typical switching frequencies. Figure 06006 shows a capacitor model that is appropriate for most applications.

The ripple due to the ESR can be approximated by first determining the current in the capacitor, assuming the capacitor to be ideal. For the buck converter in the continuous-current mode, capacitor current is the triangular current waveform of Fig. 06004c. The voltage variation across the capacitor resistance is

$$
\Delta V_{o, \mathrm{ESR}}=\Delta i_{C} r_{C}=\Delta i_{L} r_{C}
$$   [Eq. 06021]

To estimate a worst-case condition, one could assume that the peak-to-peak ripple voltage due to the ESR algebraically adds to the ripple due to the capacitance. However, __the peaks of the capacitor and the ESR ripple voltages will not coincide,__ so

$$
\Delta V_{o}<\Delta V_{o, C}+\Delta V_{o, \mathrm{ESR}}
$$   [Eq. 06022]

where $\Delta V_{o, C}$ is $\Delta V_{o}$ in [Eq. 06018]. The ripple voltage due to the ESR can be much larger than the ripple due to the pure capacitance. In that case, the output capacitor is chosen on the basis of the equivalent series resistance rather than capacitance only.

$$
\Delta V_{o} \approx \Delta V_{o, \mathrm{ESR}}=\Delta i_{C} r_{C}
$$   [Eq. 06023]

<br><br/>
<p align="center"><img src="images/capacitor_esr_model.png" style="width:39%"></p>
<center> Figure 06006.$\quad$ A model for the capacitor including the equivalent
series resistance (ESR). </center>
<br><br/>
<br><br/>

Capacitor ESR is inversely proportional to the capacitance value $-$ a larger capacitance results in a lower ESR. Manufacturers provide what are known as low ESR capacitors for power supply applications.

In -----Example 06001--------, the $100-\mu \mathrm{F}$ capacitor may have an $\mathrm{ESR}$ of $r_{C}=0.1 \Omega .$ The ripple voltage due to the ESR is calculated as

$$
\Delta V_{o, \mathrm{ESR}}=\Delta i_{C^{r} C}=\Delta i_{L} r_{C}=(1.5 \mathrm{A})(0.1 \Omega)=0.15 \mathrm{V}
$$

Expressed as a percent, $\Delta V_{o} / V_{o}$ is $0.15 / 20=0.75$ percent. The total ripple can then be approximated as 0.75 percent.

## Discontinuous-Current Operation

Continuous current in the inductor was an important assumption in the previous analyses for de-dc converters. Recall that continuous current means that the current in the inductor remains positive for the entire switching period. Continuous current is not a necessary condition for a converter to operate, but a different analysis is required for the discontinuous-current case.

__Buck Converter with Discontinuous Current__

Figure 06020 shows the inductor and source currents for discontinuous-current operation for the buck converter of Fig 06003a. 

<br><br/>
<p align="center"><img src="images/buck_converter_discontinuous_current_operation.png" style="width:39%"></p>
<center> Figure 06020.$\quad$ Buck converter discontinuous current. (a) Inductor current; (b) Source current; (c) Inductor voltage. </center>
<br><br/>
<br><br/>

The relationship between output and input voltages is determined by first recognizing that the average inductor voltage is zero for periodic operation. From the inductor voltage shown in Fig. 06020c

$$
\left(V_{s}-V_{o}\right) D T-V_{o} D_{1} T=0
$$   [Eq. ]

which is rearranged to get

$$
\left(V_{s}-V_{o}\right) D=V_{o} D_{1}
$$   [Eq. 06089]

$$
\frac{V_{o}}{V_{s}}=\left(\frac{D}{D+D_{1}}\right)
$$   [Eq. 06090]

Next, the average inductor current equals the average resistor current because the average capacitor current is zero. With the output voltage assumed constant,

$$
I_{L}=I_{R}=\frac{V_{o}}{R}
$$

Computing the average inductor current from Fig. 06020a

$$
I_{L}=\frac{1}{T}\left(\frac{1}{2} I_{\max } D T+\frac{1}{2} I_{\max } D_{1} T\right)=\frac{1}{2} I_{\max }\left(D+D_{1}\right)
$$

which results in

$$
\frac{1}{2} I_{\max }\left(D+D_{1}\right)=\frac{V_{o}}{R}
$$   [Eq. 06091]

since the current starts at zero, the maximum current is the same as the change in current over the time that the switch is closed. With the switch closed, the voltage across the inductor is

$$
v_{L}=V_{s}-V_{o}
$$

which results in

$$
\frac{d i_{L}}{d t}=\frac{V_{s}-V_{o}}{L}=\frac{\Delta i_{L}}{\Delta t}=\frac{\Delta i_{L}}{D T}=\frac{I_{\max }}{D T}
$$   [Eq. 06092]

Solving for $I_{\max }$ and using [Eq. 06089] for $\left(V_{s}-V_{o}\right) D$,

$$
I_{\max }=\Delta i_{L}=\left(\frac{V_{s}-V_{o}}{L}\right) D T=\frac{V_{o} D_{1} T}{L}
$$   [Eq. 06093]

Substituting for $I_{\max }$ in [Eq. 06091]

$$
\frac{1}{2} I_{\max }\left(D+D_{1}\right)=\frac{1}{2}\left(\frac{V_{o} D_{1} T}{L}\right)\left(D+D_{1}\right)=\frac{V_{o}}{R}
$$   [Eq. 06094]

which gives

$$ \frac{1}{2}\left(\frac{V_{o} D_{1} T}{L}\right)\left(D+D_{1}\right)-\frac{V_{o}}{R}=0 $$
$$ \frac{1}{2}\left(\frac{D_{1} T}{L}\right)\left(D+D_{1}\right)-\frac{1}{R}=0 $$
$$ \frac{T}{2 L}\left(D_{1}^{2}+D D_{1}\right)-\frac{1}{R}=0 $$
$$ D_{1}^{2}+D D_{1}-\frac{2 L}{R T}=0 $$

Solving for $D_{1}$

$$
D_{1}=\frac{-D+\sqrt{D^{2}+8 L / R T}}{2}
$$   [Eq. 06095]

Substituting for $D_{1}$ in [Eq. 06090],

$$
V_{o}=V_{s}\left(\frac{D}{D+D_{1}}\right)=V_{s}\left[\frac{2 D}{D+\sqrt{D^{2}+8 L / R T}}\right]
$$   [Eq. 06096]

The boundary between continuous and discontinuous current occurs when $D_{1}=1-D .$ Recall that another condition that occurs at the boundary between continuous and discontinuous current is $I_{\min }=0$ in [Eq. 06012].

## State Equation for The Buck Converter

State-space averaging is quite useful for developing transfer functions for switched circuits such as de-dc converters. The buck converter is used as an example. State equations for the switch closed are developed from Fig. 0B001a, and state equations for the switch open are from Fig. 0B001 b.

__Switch Closed__
First, the state equations for the buck converter (also for the forward converter) are determined for the switch closed. The outermost loop of the circuit in Fig. 0B001 a has Kirchhoff's voltage law equation

$$
L \frac{d i_{L}}{d t}+i_{R} R=V_{s}
$$   [Eq. 0B014]

Kirchhoff's current law gives

$$
i_{R}=i_{L}-i_{C}=i_{L}-C \frac{d v_{C}}{d t}
$$   [Eq. 0B015]

Kirchhoff's voltage law around the left inner loop gives

$$
L \frac{d i_{L}}{d t}+i_{C}r_{C}+v_{C}=V_{s}
$$   [Eq. 0B016]

which gives the relation

$$
i_{C}=C \frac{d v_{C}}{d t}=\frac{1}{r_{C}}\left(V_{s}-L \frac{d i_{L}}{d t}-v_{C}\right)
$$   [Eq. 0B017]

Combining [Eqs. 0B014] through [Eq. 0B017] gives the state equation

$$
\frac{d i_{L}}{d t}=-\frac{R r_{C}}{L\left(R+r_{C}\right)} i_{L}-\frac{R}{L\left(R+r_{C}\right)} v_{C}+\frac{1}{L} V_{s}
$$   [Eq. 0B018]

Kirchhoff's voltage law around the right inner loop gives

$$
-v_{C}-i_{C} r_{C}+i_{R} R=0
$$   [Eq. 0B019]

Combining the above equation with [Eq. 0B015] gives the state equation

$$
\frac{d v_{C}}{d t}=\frac{R}{C\left(R+r_{C}\right)} i_{L}-\frac{1}{C\left(R+r_{C}\right)} v_{C}
$$   [Eq. 0B020]

Restating [Eq. 0B018] and [Eq. 0B020] in state-variable form gives

$$
\dot{x}=A_{1} x+B_{1} V_{s}
$$   [Eq. 0B021]

where

$$
\dot{x}=\left[\begin{array}{l}
\dot{i}_{L} \\
\dot{v}_{C}
\end{array}\right]
$$

$$
A_{1}=\left[\begin{array}{cc}
-\frac{R r_{C}}{L\left(R+r_{C}\right)} & -\frac{R}{L\left(R+r_{C}\right)} \\
\frac{R}{C\left(R+r_{C}\right)} & -\frac{1}{C\left(R+r_{C}\right)}
\end{array}\right]
$$   [Eq. 0B022]

$$
B_{1}=\left[\begin{array}{l}
\frac{1}{L} \\
0
\end{array}\right]
$$

If $r_{C}<<R$

$$
A_{1} \approx\left[\begin{array}{cc}
-\frac{r_{C}}{L} & -\frac{1}{L} \\
\frac{1}{C} & -\frac{1}{R C}
\end{array}\right]
$$   [Eq. 0B023]

__Switch Open__
The filter is the same for the switch closed as for the switch open. Therefore, the $A$ matrix remains unchanged during the switching period.

$$
A_{2}=A_{1}
$$

The input to the filter is zero when the switch is open and the diode is conducting. State equation [Eq. 0B016] is modified accordingly, resulting in

$$
B_{2}=0
$$

Weighting the state variables over one switching period gives

$$
\begin{array}{c}
\dot{x} d=A_{1} x d+B_{1} V_{s} d \\
\dot{x}(1-d)=A_{2} x(1-d)+B_{2} V_{s}(1-d)
\end{array}
$$   [Eq. 0B024]

Adding the above equations and using $A_{2}=A_{1}$,

$$
\dot{x}=A_{1} x+\left[B_{1} d+B_{2}(1-d)\right] V_{s}
$$   [Eq. 0B025]

In expanded form,

$$
\left[\begin{array}{c}
i_{L} \\
\dot{v}_{C}
\end{array}\right]=\left[\begin{array}{cc}
-\frac{r_{C}}{L} & -\frac{1}{L} \\
\frac{1}{C} & -\frac{1}{R C}
\end{array}\right]\left[\begin{array}{c}
i_{L} \\
v_{C}
\end{array}\right]+\left[\begin{array}{c}
\frac{d}{L} \\
0
\end{array}\right] V_{s}
$$   [Eq. 0B026]

[Eq. 0B026] gives the averaged state-space description of the output filter and load of the forward converter or buck converter. 

__The output voltage $v_{o}$__ is determined from

$$
v_{o}=R i_{R}=R\left(i_{L}-i_{R}\right)=R\left(i_{L}-\frac{v_{o}-v_{C}}{r_{C}}\right)
$$   [Eq. 0B027]

Rearranging to solve for $v_{o}$

$$
v_{o}=\left(\frac{R r_{C}}{R+r_{C}}\right) i_{L}+\left(\frac{R}{R+r_{C}}\right) v_{C} \approx r_{C} i_{L}+v_{C}
$$   [Eq. 0B028]

The above output equation is valid for both switch positions, resulting in $C_{1}^{T}=$ $C_{2}^{T}=C^{T} .$ In state-variable form

$$
v_{o}=C^{T} x
$$

where

$$
C^{T}=\left[\frac{R r_{C}}{R+r_{C}} \quad \frac{R}{R+r_{C}}\right] \approx\left[r_{C} \quad 1\right]
$$   [Eq. 0B029]

and

$$
x=\left[\begin{array}{l}
i_{L} \\
v_{C}
\end{array}\right]
$$   [Eq. 0B030]

The steady-state output is found from [Eq. 0B009],

$$
V_{o}=-C^{T} A^{-1} B V_{s}
$$   [Eq. 0B031]

where $A=A_{1}=A_{2}, B=B_{1} D,$ and $C^{T}=C_{1}^{T}=C_{2}^{T} .$ The final result of this computation results in a steady-state output of

$$
V_{o}=V_{s} D
$$   [Eq. 0B032]

__The small-signal__ transfer characteristic is developed from [Eq. 0B012], which in the case of the buck converter results in

$$
\stackrel{\leftrightarrow}{x}=A \widetilde{x}+B V_{s} \tilde{d}
$$   [Eq. 0B033]

Taking the Laplace transform,

$$
s \tilde{x}(s)=A \tilde{x}(s)+B V_{s} \tilde{d}(s)
$$   [Eq. 0B034]

Grouping $\tilde{x}(s)$

$$
[s I-A] \tilde{x}(s)=B V_{s} \tilde{d}(s)
$$   [Eq. 0B035]

where $I$ is the identity matrix. Solving for $\tilde{x}(s)$,

$$
\tilde{x}(s)=[s I-A]^{-1} B V_{s} \tilde{d}(s)
$$   [Eq. 0B036]

Expressing $\tilde{v}_{o}(s)$ in terms of $\tilde{x}(s)$

$$
\tilde{v}_{o}(s)=C^{T} \tilde{x}(s)=C^{T}[s I-A]^{-1} B V_{s} \tilde{d}(s)
$$   [Eq. 0B037]

Finally, the transfer function of output to variations in the duty ratio is expressed as

$$
\frac{\tilde{v}_{o}(s)}{d(s)}=C^{T}[s I-A]^{-1} B V_{s}
$$   [Eq. 0B038]

Upon substituting for the matrices in the above equation, a lengthy evaluation process results in the transfer function

$$
\frac{\tilde{v}_{o}(s)}{\tilde{d}(s)}=\frac{V_{s}}{L C}\left[\frac{1+s r_{C} C}{s^{2}+s\left(1 / R C+r_{C} / L\right)+1 / L C}\right]
$$   [Eq. 0B039]

The above transfer function will used in the section on [07000 - Control of dc power supplies]

## Buck Converter Design Considerations

Most buck converters are designed for continuous-current operation. The choice of switching frequency and inductance to give continuous current is given by [Eq. 06013], and the output voltage ripple is described by [Eq. 06016] and [Eq. 06021]. Note that as the switching frequency increases, the minimum size of the inductor to produce continuous current and the minimum size of the capacitor to limit output ripple both decrease. Therefore, high switching frequencies are desirable to reduce the size of both the inductor and the capacitor.

The tradeoff for high switching frequencies is increased power loss in the switches. Increased power loss in the switches means that heat is produced. This decreases the converter's efficiency and may require a large heat sink, offsetting the reduction in size of the inductor and capacitor. Typical switching frequencies are above $20 \mathrm{kHz}$ to avoid audio noise, and they extend well into the $100 \mathrm{s}$ of kilohertz and into the megahertz range. Some designers consider about $500 \mathrm{kHz}$ to be the best compromise between small component size and efficiency. Other designers prefer to use lower switching frequencies of about $50 \mathrm{kHz}$ to keep switching losses small, while still others prefer frequencies larger than $1 \mathrm{MHz}$. As switching devices improve, switching frequencies will increase.

For low-voltage, high-current applications, the synchronous rectification scheme of Fig. 06007 is preferred over using a diode for the second switch. The voltage across the conducting MOSFET will be much less than that across a diode, resulting in lower losses.

The inductor value should be larger than $L_{\min }$ in [Eq. 06014] to ensure continuouscurrent operation. Some designers select a value 25 percent larger than $L_{\min }$. Other designers use different criteria, such as setting the inductor current variation, $\Delta i_{L}$ in [Eq. 06015], to a desired value, such as $40$ percent of the average inductor current. A smaller $\Delta i_{L}$ results in lower peak and $\mathrm{rms}$ inductor currents and a lower rms capacitor current but requires a larger inductor.

The inductor wire must be rated at the rms current, and the core should not saturate for peak inductor current. The capacitor must be selected to limit the output ripple to the design specifications, to withstand peak output voltage, and to carry the required rms current.

The switch (usually a MOSFET with a low $R_{D S \text { on }}$ ) and diode (or second MOSFET for synchronous rectification) must withstand maximum voltage stress when off and maximum current when on. The temperature ratings must not be exceeded, often requiring a heat sink.

Assuming ideal switches and an ideal inductor in the initial design is usually reasonable. However, the ESR of the capacitor should be included because it typically gives a more significant output voltage ripple than the ideal device and greatly influences the choice of capacitor size.

## __Synchronous Rectification for the Buck Converter__
Many buck converters use a second MOSFET in place of the diode. When $S_{2}$ is on and $S_{1}$ is off, current flows upward out of the drain of $S_{2}$. The advantage of this configuration is that the second MOSFET will have a much lower voltage drop across it compared to a diode, resulting in higher circuit efficiency. This is especially important in low-voltage, high-current applications. A Shottky diode would have a voltage of 0.3 to $0.4 \mathrm{V}$ across it while conducting, whereas a MOSFET will have an extremely low voltage drop due to an $R_{D S_{\mathrm{on}}}$ as low as single-digit milliohms. This circuit has a control scheme known as synchronous switching, or synchronous rectification. The second MOSFET is known as a synchronous rectifier. The two MOSFETs must not be on at the same time to prevent a short circuit across the source, so a "dead time" is built into the switching control one MOSFET is tumed off before the other is turned on. A diode is placed in parallel with the second MOSFET to provide a conducting path for inductor current during the dead time when both MOSFETs are off. This diode may be the MOSFET body diode, or it may be an extra diode, most likely a Shottky diode, for improved switching. The synchronous buck converter should be operated in the continuous-current mode because the MOSFET would allow the inductor current to go negative.

All the DC DC converter topologies presented in this and the the upcoming sections, can utilize MOSFETs in place of diodes.

# Nonideal Switches and Buck Converter Performance

Switch Voltage Drops
All the preceding calculations were made with the assumption that the switches were ideal. Voltage drops across conducting transistors and diodes may have a significant effect on converter performance, particularly when the input and output voltages are low. Design of de-de converters must account for nonideal components. The buck converter is used to illustrate the effects of switch voltage drops.

Referring again to the analysis of the buck converter of Fig. 06003a, the inputoutput voltage relationship was determined using the inductor voltage and current. With nonzero voltage drops across conducting switches, the voltage across the inductor with the switch closed becomes

$$
v_{L}=V_{s}-V_{o}-V_{Q}
$$   [Eq. 06086]

where $V_{Q}$ is the voltage across the conducting switch. 

With the switch open, the voltage across the diode is $V_{D}$ and the voltage across the inductor is

$$
v_{L}=-V_{o}-V_{D}
$$   [Eq. 06087]

The average voltage across the inductor is zero for the switching period.

$$
V_{L}=\left(V_{s}-V_{o}-V_{Q}\right) D+\left(-V_{o}-V_{D}\right)(1-D)=0
$$

Solving for $V_{o}$

$$
V_{o}=V_{s} D-V_{Q} D-V_{D}(1-D)
$$   [Eq. 06088]

which is lower than $V_{o}=V_{s} D$ for the ideal case.

## Switching Losses

In addition to the on-state voltage drops and associated power losses of the switches, other losses occur in the switches as they turn on and off. Figure 06019a illustrates switch on-off transitions. For this case, it is assumed that the changes in voltage and current are linear and that the timing sequence is as shown. The instantaneous power dissipated in the switch is shown in Fig. 06019a. Another possible switch on-off transition is shown in Fig. 06012b. In this case, the voltage and current transitions do not occur simultaneously. This may be closer to actual switching situations, and switching power loss is larger for this case.

The energy loss in one switching transition is the area under the power curve. Since the average power is energy divided by the period, higher switching frequencies result in higher switching losses. One way to reduce switching losses is to modify the circuit to make switching occur at zero voltage and/or zero current. This is the approach of The Resonant Power Converter.
